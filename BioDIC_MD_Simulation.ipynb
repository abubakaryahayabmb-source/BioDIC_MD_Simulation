```python
# Select 1 (Protein) for centering and 0 (System) for output
!printf "1\n0\n" | gmx trjconv -s MD.tpr -f final_100ns.xtc -o MD_center.xtc -pbc mol -center
```

```python
# 1. RMSD (Protein Backbone and Ligand)
!printf "4\n4\n" | gmx rms -s MD.tpr -f MD_center.xtc -o rmsd_prot.xvg -tu ns
!printf "4\n13\n" | gmx rms -s MD.tpr -f MD_center.xtc -o rmsd_lig.xvg -tu ns -n index.ndx

# 2. RMSF (Per-residue fluctuation)
!printf "3\n" | gmx rmsf -s MD.tpr -f MD_center.xtc -o rmsf.xvg -res

# 3. Radius of Gyration (Rg)
!printf "1\n" | gmx gyrate -s MD.tpr -f MD_center.xtc -o gyrate.xvg

# 4. Hydrogen Bonds (Protein-Ligand)
!printf "1\n13\n" | gmx hbond -s MD.tpr -f MD_center.xtc -num hbond.xvg -n index.ndx

# 5. SASA (Solvent Accessible Surface Area)
!printf "1\n" | gmx sasa -s MD.tpr -f MD_center.xtc -o sasa_prot.xvg
!printf "13\n" | gmx sasa -s MD.tpr -f MD_center.xtc -o sasa_lig.xvg -n index.ndx
```

```python
import matplotlib.pyplot as plt
import pandas as pd

def plot_xvg(file, title, xlabel, ylabel, label=None):
    data = pd.read_csv(file, sep='\s+', comment='@', header=None, names=['x', 'y'])
    plt.plot(data['x'], data['y'], label=label)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)

# Plot 1: RMSD (Protein vs Ligand)
plt.figure(figsize=(10,5))
plot_xvg('rmsd_prot.xvg', 'RMSD over Time', 'Time (ns)', 'RMSD (nm)', label='Protein')
plot_xvg('rmsd_lig.xvg', 'RMSD over Time', 'Time (ns)', 'RMSD (nm)', label='Ligand')
plt.legend()
plt.show()

# Plot 2: Hydrogen Bonds
plt.figure(figsize=(10,4))
plot_xvg('hbond.xvg', 'Protein-Ligand Hydrogen Bonds', 'Time (ns)', 'Number of HBonds')
plt.show()

# Plot 3: Radius of Gyration
plt.figure(figsize=(10,4))
plot_xvg('gyrate.xvg', 'Radius of Gyration (Rg)', 'Time (ns)', 'Rg (nm)')
plt.show()
```
